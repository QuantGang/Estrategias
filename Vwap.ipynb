Calculate VWAP

```
import matplotlib.pyplot as plt
import pandas as pd

Rollsize = 7
VWc = pd.DataFrame((mydata['High']+mydata['Low']+mydata['Close'])/3, columns=['price'])
VWc['Volprice'] = VWc['price']*mydata['Volume']
VWc['VWap'] = VWc['Volprice'].rolling(window=Rollsize).sum()/mydata['Volume'].rolling(window=Rollsize).sum()
mydata['Adj Close'].plot()
VWc['VWap'].plot()
plt.plot()
```
