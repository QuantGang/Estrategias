La función momentum_pos, recibe cuatro parámetros
1. El ticker del activo
2. El tamaño de la ventana a trabajar.
3. El tamaño de la ventana de la media móvil.
4. La fecha en la que se quiere obtener la recomendación de inversión.

El programa devuelve $1$ para posición larga $-1$ para posición corta.

```
mm.momentum_pos('BTC-USD',180,30)
```

La función momentum_pos_noss, es igual a la función momentum_pos pero no permite venta corta por lo que devuelve $1$ para posición larga $0$ para posición de no tener el activo (solo cash).

```
mm.momentum_pos_noss('BTC-USD',180,30)
```

La función back_test_momentum_returns, recibe cinco parámetros:
1. Fecha de inicio del backtest.
2. Fecha de finalización del backtest.
3. Ticker
4. Tamaño de la ventana de trabajo.
5. Tamaño de la ventana móvil.

Imprime unos gráficos representativos del desempeño del activo.

```
mm.back_test_momentum_returns(datetime(2019,10,1), datetime(2020,2,1), 'BTC-USD', 180, 90)
```

La función back_test_momentum_capital es igual a la función back_test_momentum_returns, pero trabaja con el capital del inversionista e imprime el capital inicial y final del backtest.

```
mm.back_test_momentum_capital(datetime(2019,10,1), datetime(2019,12,1), 'BTC-USD', 120, 30,100)
```
